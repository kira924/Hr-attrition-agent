```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
pd.set_option('display.max_columns', None)
```

```python
data_path = "../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv"
```

```python
try:
    df = pd.read_csv(data_path)
    print(" Data Loaded Successfully!")
    print(f"Shape: {df.shape}")  
except FileNotFoundError:
    print(" Error: File not found. Please check the path.")
```

```python
display(df.head())
```

```python
print("\nTarget Distribution:")
print(df['Attrition'].value_counts(normalize=True))
```

```python
print("\nMissing Values:")
print(df.isnull().sum().sum())
```

```python
drop_cols = ['EmployeeCount', 'Over18', 'StandardHours']
df.drop(columns=drop_cols, inplace=True, errors='ignore')
print(f"\nDropped columns: {drop_cols}")
print(f"New Shape: {df.shape}")
```

```python
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/01_preprocessed.csv', index=False)
print("\n Preprocessed data saved to data/processed/01_preprocessed.csv")
```
